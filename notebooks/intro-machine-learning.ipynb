# An Introduction to Machine Learning
## A taxonomy of learning approaches



### What is machine learning?

The field of machine learning (ML) seeks to answer the question "How can we build systems that improve with data, and what are the fundamental laws that govern learning processes?"


ML blends principles of computer science and statistics to create sophisticated algorithms. ML can be used to:
  - generate new data (generative learning), 
  - identify groupings (clustering), 
  - predict new values for unseen observations (regression), 
  - predict labels for unseen observations(classification), 
  - complete a task (reinforcement learning),
  - or make recommendations (information filtering).

The common theme for ML is automation. We want to know how we can automate processes (prediction, clustering, generation, recommending, task completion) by learning directly from data.

### What do we mean by **machine** and **learning**?

- We need an algorithm with instructions to create a model that learns from data.
- This is different from a rule based model where rules are hard coded by humans (if-else statements for instance).
- To make accurate predictions, a model requires data that holds information the model can use, often called features.
- A prediction might be _how much it will rain_, or _if person X will catch a bus tomorrow_.
- Complex machine learning models are used to identify relationship between features, and a desired output.

### The brief history of Machine Learning (ML)

- The field of machine learning has its roots in the 1950s, with early research on artificial intelligence and pattern recognition. 

- Pioneers in the field include Alan Turing, Claude Shannon, and Norbert Wiener. 

- Today, some notable pioneers in the field of machine learning include Geoffrey Hinton, Yann LeCun, and Yoshua Bengio, who have made significant contributions to the development of deep learning.

### Supervised learning vs unsupervised learning:

**Supervised learning** involves training an algorithm on a labeled dataset, where **labelled** means we have example answers for our input features. The goal is to then make new predictions given new input features. The labeled examples provide the algorithm with information on the correct output for a given input. eg:

- Regression (prediction of continuous numerical values) - _What will be the stock value in one year?_
- Classification (prediction of a categorical class label) - _Will the stock value increase or decrease?_

**Unsupervised learning** is a type of machine learning where the algorithm is trained on an unlabeled dataset and tries to identify patterns or relationships in the data without any prior knowledge of the correct output. The goal is to understand the underlying structure of the data. eg:

- Clustering (grouping objects based on the similarity of their attributes) - _Are there different types of travel behavior?_



**Other ML approaches**:

- **Reinforcement learning** is a type of ML without given labels. Instead the model learns how to best behave in order to achieve some goal in a given environment. For example the model may learn to play Chess.
- **Self-supervised learning** is a type of learning where a task is devised that uses the input data to create labels. For example, given some input text, random words can be masked (hidden), then the model asked to predict them. Solving these tasks allows the model to learn underlying structures in the data. This strategy is a common foundation in modern large language models (LLMs).
- **Generative** is a type of learning, typically considered as self-supervised but where the objective is to learn the distribution of the data such that new samples can be generated.


### Machine learning tasks

Example tasks where ML can be deployed:
  - **Predicting** house prices (Forecasting trends)
  - **Classifying** document topics (Natural Language Processing)
  - **Grouping** similar users in social networks (Unsupervised clustering)

ML algorithms are widely used in many real-world applications and have proven to be effective in solving a variety of problems. Understanding ML algorithms requires one to understand their limitations and which methods, practices and models to choose.


#### Deep learning versus "Shallow"

Deep learning typically refers to a particular type of model with multiple layers. In theory these layers allow "deep" models to learn and use more complex patterns. These models are often described as "Neural Networks". Most modern advancements in computer vision and language modelling use extremely deep models.

Additional deep learning considerations:

- **Feature Engineering**: Some traditional ML algorithms necessitate extensive feature engineering, where relevant features need to be extracted from raw data. Deep learning algorithms, on the other hand, can automatically learn useful features from data, without manual selection of input features or an exploratory data analysis.

- **Large Amount of Data**: Deep learning algorithms perform better with larger amounts of data, which might not be the case with more traditional ML algorithms.

- **Non-linear Relationships**: Deep learning algorithms model non-linear relationships between inputs and outputs, while traditional ML algorithms might not include non linear transformations of the input to draw non linear relationships with the output.

- **Computational Requirements**: Deep learning algorithms are computationally intensive, requiring powerful GPUs for training, while traditional ML algorithms are typically less computationally demanding.

There is no common term for non-deep ML, but such smaller models still have plenty of use for the right problems.
Examples of non-deep models include the perceptron, logistic classification, linear regression, etc.
They are typically suitable for smaller, well-defined problems with a limited amount of data. Whilst deep learning algorithms are well suited for more complex, large-scale problems with unstructured data.

### A typical ML pipeline



It is important to remember that training a model is only one part of a broader pipeline.

### Supervised Learning

- Data includes **input** features $X$ and a **label** $Y$ (sometimes called **observed** or **target**).
- Where there is some process $f$ generating Y from X, such that $Y = f(X)$.
- These processes can be quite abstract in nature, consider for example the process that generates our salary (Y) based on inputs (X) such as years in education.
- Or the process could be a physical one such as the global weather system.
- Note that $f(X)$ is ussually considered a stochastic process - such that $Y$ is drawn from a distribution.
- The goal of a supervised learning model is to approximate the data generation function to make predictions $\hat{Y}$ with:

    $\hat{Y} = f'(X)$
   

#### Examples of a supervised learning problems:

- **Predicting house prices** - features could include: number of rooms, postcode, etc.
- **Predicting the probability a person has diabetes** - features could include: blood pressure, age, etc.

### Unsupervised Learning

- Data includes **input** features $X$ with no labels.
- Goal is to find structure in the **input** data $X$
- For example to **group** similar data points (**clustering**)
- Or reduce the number of **data** dimensions (dimensionality reduction)

### Examples of unsupervised learning problems:

- **Customer clustering** - given data about customer buying habits, identify the primary different types of purchasing habits.
- **Dimensionality Reduction** - you receive a large survey dataset describing individuals' responses to 100 questions about their lives, reduce the complexity of this dataset down to only three dimensions but maintaining as much as the original information as possible.

### Choosing a model

There are many factors that indicate the most appropriate model:
  - **Data type**: tabular, image, time series, text
  - **Data properties**: number of dimensions, number of samples, missing values
  - **Task**: prediction, clustering, visualization, control
  - **High-level goals**: interpretability, performance, robustness, computational load

An important part of modeling expertise is to be familiar with the broad range of ML models, and which to choose for each problem. There is a large number of available models, each with their strengths and weaknesses. In this course we cover many models, with examples of when they can be useful.

### Model zoo



For each task and type of data, there are multiple model choices. Part of ML expertise is knowing which to use and how to tune its hyper-parameters and train it. Libraries make it easy to try multiple models with your data.

### What to achieve from the data? Generalisation!

In supervised learning we want to use the available data to:
  - (1) learn a predictive model
  - (2) assess the model's performance
    
To check for generalisation we **train** the model on a training dataset and **assess** the model empirically on a test set. We split the data into two parts:
  - Data for (1): **training set**
  - Data for (2): **test set**
    
In addition to the training and test set, it is often sensible to also use a **validation** set to allow us to experiment with different models and hyper-parameters without over-fitting (more on this later).


### How does a model generalise to new data?
    
   
- After training a model usually performs well on the training data.
- A fundamental concept in machine learning is that a model should also perform well on **unseen data**, i.e. the test set.
- A model that performs well on the training data but not on the test data has failed to gereralise.
- This is also called over-fitting, ie over-fitting the model to the training data.
- Our data has different levels of **complexity** and our models of **capacity**.
- If we throw too much capacity at too little complexity, our models will effectively memorise the training data and might fail to generalise to the test data.

### Bias versus Variance

- More formally, the problem or under or over-fitting is caused by a trade-off between model **Bias** and **Variance**.
- The **bias** is an error from incorrect or simplifying assumptions in the model. Models want to reduce bias. Models with high bias fail to learn the relations between input features and target outputs (under-fitting). But reducing bias, for example by increasing model capacity/complexity, typically increases variance. 
- The **variance** is an error from over-sensitivity to small fluctuations in the training set. This results in the model learning patterns in the training data that do not generalise to the test data (over-fitting).



### Complex models
 
Modern machine learning, in particular **deep learning**, expands modeling to a **very large** number of parameters, dimensions and data entries.



An important difference between models is their complexity. Traditional ML models are usually smaller, sometimes linear, with strong assumptions about the data, like Gaussian clusters. 

Other models such as deep neural networks have many parameters, are highly nonlinear and model high-dimensional data.

### Data and computing power

Larger models with more capacity require more data and more computing power to train. Coincidentally, a lot of data on many subjects has become available over the past few decades, driven by the expansion of computing systems and the internet.

For machine learning, particularly deep learning, the development of graphics cards (GPUs) has significantly accelerated development in the field.



### Deep learning revolution


- Deep neural networks have scaled machine learning to achieve state of the art results in a wide range of applications.
- Driven by large computational resources and datasets.
- Neural networks are the reference model for many domains, including images, language, and audio.
- State-of-the-art methods allow processing highly-dimensional multi-modal (of different types) data of arbitrary distributions which cannot be analysed using conventional statistics.
- Common examples include geospatial data, multi-dimensional time series, and social networks data.
- Neural networks are extremely flexible function approximators, given enough data, neural networks can approximate a wide range of functions.
- It might also be useful to think of neural networks as compression algorithms, able to extract and store critical information from large datasets.



Traditional ML is being supplanted by neural networks for large complex datasets.
